```python
df=df.drop(columns=["Unnamed: 13","Unnamed: 0"])
```

```python
df[["Inns","NO","Runs","HS","Ave","BF","SR","100","50","0"]]=df[["Inns","NO","Runs","HS","Ave","BF","SR","100","50","0"]].replace("-",0)
```

```python
df.info()
```

```python
df.columns
```

```python
df[['Inns', 'NO', 'Runs','100', '50', '0', 'BF']]=df[['Inns', 'NO', 'Runs','100', '50', '0', 'BF']].astype(int)
df[['SR','Ave']] = df[['SR',"Ave"]].astype(float)
df[["Player","Span","HS"]]=df[["Player","Span","HS"]].astype(str)
```

```python
df.info()
```

```python
#user defined function to check whether batter was not out at their highest score
def check_no(value):
    if "*" in value:
        return 1
    else:
        return 0
```

```python
df["HS_no"] = df["HS"].apply(check_no)
```

```python
#user defined function to remove "*" - which represents not out
def cleanHS(value):
    if "*" in value:
        return value[:-1]
    else:
        return value
```

```python
df["HS"]=df["HS"].apply(cleanHS)
df["HS"]=df["HS"].astype('int')
```

```python
df.head()
```

```python
#splitting debut and retirement dates
df[["debut","retire"]]=df["Span"].str.split('-',expand=True).astype('int')
df.drop(columns="Span",inplace=True)
```

```python
df.head()
```

```python
#direct method to split player name and country
df["Player"]=df["Player"].str.replace("(",'/')
df["Player"]=df["Player"].str.replace(")",'')
df[["Player","team1","team2","team3"]]=df["Player"].str.split("/",expand=True)

for col in df[["Player","team1","team2","team3"]]:
    df[col] = df[col].apply(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
df["team1"]=df["team1"].replace('3',None)
df["team1"]=df["team1"].replace('1',None)
```

```python
df = df.rename(columns={"Inns":"Inns_bat",
               "Runs":"Runs_scored",
               "Ave":"Avg_bat",
               "SR":"SR_bat"
                       })
```

```python
df["pid"] = df["Player"]+df["debut"].astype("str")+df["retire"].astype("str")
df["pid"] = df["pid"].str.replace(" ",'')

df = df[['pid','Player', 'Mat', 'Inns_bat', 'NO', 'Runs_scored', 'HS', 'Avg_bat', 'BF',
       'SR_bat', '100', '50', '0', 'HS_no', 'debut', 'retire', "team1","team2","team3"]]
df.head(3)
```
