```python
df[["BBwkts","BBruns"]]=df["BBI"].str.split("/",expand=True)
df.head()
```

```python
df[["BBwkts","BBruns"]]=df[["BBwkts","BBruns"]].fillna(0)
```

```python
df = df.drop(columns=["BBI"],axis=1)
```

```python
df.head()
```

```python
#direct method to split player name and country
df["Player"]=df["Player"].str.replace("(",'/')
df["Player"]=df["Player"].str.replace(")",'')
df[["Player","team1","team2","team3"]]=df["Player"].str.split("/",expand=True)

for col in df[["Player","team1","team2","team3"]]:
    df[col] = df[col].apply(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
df["team1"]=df["team1"].replace('3',None)
df["team1"]=df["team1"].replace('1',None)
```

```python
df.head()
```

```python
df[['Inns', 'Balls', 'Runs', 'Wkts', '4', '5','debut', 'retire','BBwkts', 'BBruns']]=df[['Inns', 'Balls', 'Runs', 'Wkts', '4', '5','debut', 'retire','BBwkts', 'BBruns']].astype('int')
df[['Ave', 'Econ', 'SR']]=df[['Ave', 'Econ', 'SR']].astype('float')
```

```python
df= df[['Player', 'Mat', 'Inns', 'Balls', 'Runs', 'Wkts', 'Ave', 'Econ', 'SR',
       '4', '5', 'BBwkts', 'BBruns', 'debut', 'retire', "team1","team2","team3"]]
```

```python
df.columns
```

```python
df = df.rename(columns={"Inns":"Inns_bowl",
               "Balls":"Balls_bowled",
               "Runs":"Runs_given",
               "Wkts":"Wkts_taken",
               "Ave":"Avg_bowl",
               "Econ":"Econ_bowl",
               "SR":"SR_bowl",
               "4":"4wkt",
               "5":"5wkt"})
```

```python
df["pid"] = df["Player"]+df["debut"].astype("str")+df["retire"].astype("str")
df["pid"] = df["pid"].str.replace(" ",'')

df = df[[ 'pid','Player', 'Mat', 'Inns_bowl', 'Balls_bowled', 'Runs_given',
       'Wkts_taken', 'Avg_bowl', 'Econ_bowl', 'SR_bowl', '4wkt', '5wkt',
       'BBwkts', 'BBruns', 'debut', 'retire',"team1","team2","team3"]]
df.head(3)
```
